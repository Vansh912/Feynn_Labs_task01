```python
#as these columns does not contain binary output we will remove them 
data_new = data.drop(labels=['Like','Age','VisitFrequency','Gender'], axis=1) 
data_new.head()
```

```python
label_encoder = preprocessing.LabelEncoder()
data_new1= data_new.apply(label_encoder.fit_transform)
data_new1
```

```python
pca = PCA(n_components=11)

principalComponents = pca.fit_transform(data_new1)

principalDf = pd.DataFrame(data = principalComponents, columns = ['yummy','convenient','spicy','fattening','greasy','fast','cheap','tasty','expensive','healty','disgusting'])
principalDf.head()
```

```python
pca.explained_variance_ratio_
```

```python
loadings = pca.components_
num_pc = pca.n_features_
pc_list = ["PC"+str(i) for i in list(range(1, num_pc+1))]
loadings_df = pd.DataFrame.from_dict(dict(zip(pc_list, loadings)))
loadings_df['feature'] = data_new1.columns.values
loadings_df = loadings_df.set_index('feature')
loadings_df
```

```python
wcss = []

for k in range(2,15):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data_new1)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(15,5))
plt.plot(range(2,15),wcss)
plt.xlabel("number of k (cluster) value")
plt.ylabel("wcss")
plt.show()
```

```python
K = range(2,12)
wcss=[]
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(data_new1)
    wcss.append(kmeanModel.inertia_)
plt.figure(figsize=(7,5))
plt.plot(range(2,12),wcss)
plt.xlabel("number of k (cluster) value")
plt.ylabel("wcss")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=0)
kmeans.fit(data_new1)
print('No. of iterations: ', kmeans.n_iter_)   
```
